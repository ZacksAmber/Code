```python
import pandas as pd
```

```python
# Define cones' attributes
conesAttributes = ["Flavor", "Color", "Price"]
```

```python
# Define cones' data, must be same sequence as attributes
conesData = [["strawberry", "pink", 3.55],
        ["cholocate", "light brown", 4.75],
        ["cholocate", "dark brown", 5.25],
        ["strawberry", "pink", 5.25],
        ["cholocate", "dark brown", 5.25],
        ["bubblegem", "pink", 4.75]]
```

```python
# Define cones
cones = pd.DataFrame(data=conesData, columns=conesAttributes)
```

```python
# Show cones
cones
```

```python
# Show first two rows
cones.iloc[0:2, :]
```

```python
# Show Flavor
cones.loc[:, ['Flavor']]
```

```python
# Show Flavor, Price
cones.loc[:, ['Flavor', 'Price']]
```

```python
# Drop Color
cones.drop(columns='Color')
```

```python
# Redefine cones
cones = pd.DataFrame(data=conesData, columns=conesAttributes)

# Name a new dataframe without Color
no_colors = cones.drop(columns='Color')
```

```python
# Show no_colors
no_colors
```

```python
# Sort Price
cones.sort_values(by=["Price"])
```

```python
# Sort Price Descending
cones.sort_values(by=["Price"], ascending=False)
```

```python
# Making boolean series for a cones
filter = cones['Flavor'] == "cholocate"

# Pass filter to cones
cones.loc[filter]
```

```python
# Making boolean series for a cones
filter = cones['Flavor'] == "cholocate"

# Pass filter to cones
cones.where(filter, inplace=False)
```

```python
# Making boolean series for a cones
filter = cones['Flavor'] == "Cholocate"

# Pass filter to cones
cones.loc[filter]
```
